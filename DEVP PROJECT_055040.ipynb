```python
non_cat = ['Quantity', 'Value', 'Customs_Code', 'Weight','Invoice_Number']

# Create boxplots for each non-categorical variable with mean line
for i in non_cat:
    plt.figure()
    plt.boxplot(my_sample[i])
    plt.title(f'Boxplot of {i}')
  
  # Calculate and plot the mean line
    mean_value = my_sample[i].mean()
    plt.axhline(mean_value, color='r', linestyle='--', label='Mean')
    plt.legend()
```

```python
# Categorical variables
cat_var = ['Country', 'Product', 'Import_Export', 'Category', 'Port', 'Shipping_Method', 'Supplier','Customer','Payment_Terms']

# Generate bar graphs and pie charts for each categorical variable
for var in cat_var:
    value_counts = my_sample[var].value_counts()

  # Bar graph
    plt.figure()
    plt.bar(value_counts.index, value_counts.values)
    plt.title(f'Bar Graph of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')

  # Pie chart
    plt.figure()
    plt.pie(value_counts.values, labels=value_counts.index, autopct='%1.1f%%')
    plt.title(f'Pie Chart of {var}')

plt.show()
```

```python
#Transaction Value Distribution : Histogram showing the distribution of transaction values.

plt.figure(figsize=(10,6))
sns.histplot(my_sample['Value'], bins=30, kde=True)
plt.title('Distribution of Transaction Values')
plt.xlabel('Transaction Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#Top 10 Countries by Export Volume : Bar plot showing the top 10 countries with the highest export quantities.

top_countries = my_sample[my_sample['Import_Export'] == 'Export'].groupby('Country')['Quantity'].sum().nlargest(10)
top_countries.plot(kind='bar', figsize=(10,6), color='skyblue')
plt.title('Top 10 Countries by Export Volume')
plt.xlabel('Country')
plt.ylabel('Total Quantity Exported')
plt.xticks(rotation=45)
plt.show()
```

```python
#Imports vs Exports by Month : Line plot comparing import and export values over time (monthly)

my_sample['Date'] = pd.to_datetime(my_sample['Date'])
my_sample['Month'] = my_sample['Date'].dt.to_period('M')

monthly_data = my_sample.groupby(['Month', 'Import_Export'])['Value'].sum().unstack()
monthly_data.plot(kind='line', figsize=(10,6), marker='o')
plt.title('Imports vs Exports Over Time')
plt.xlabel('Month')
plt.ylabel('Transaction Value')
plt.show()
```

```python
#Shipping Methods Comparison : Pie chart of shipping methods used across all transactions.

shipping_method_counts = my_sample['Shipping_Method'].value_counts()
shipping_method_counts.plot(kind='pie', autopct='%1.1f%%', figsize=(8,8), startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Shipping Methods')
plt.ylabel('')
plt.show()
```

```python
#Transaction Value by Category : Box plot showing the distribution of transaction values for different product 

plt.figure(figsize=(12,6))
sns.boxplot(x='Category', y='Value', data=my_sample)
plt.title('Transaction Value Distribution by Product Category')
plt.xticks(rotation=45)
plt.show()
```

```python
#Country-wise Import vs Export Comparison: Stacked bar plot showing country-wise imports vs exports.

country_data = my_sample.groupby(['Country', 'Import_Export'])['Value'].sum().unstack()
country_data.nlargest(10, 'Export').plot(kind='bar', stacked=True, figsize=(12,6))
plt.title('Country-wise Import vs Export Comparison')
plt.xlabel('Country')
plt.ylabel('Transaction Value')
plt.xticks(rotation=45)
plt.show()
```
